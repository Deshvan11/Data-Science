**MERGING TWO DATAFRAMES**

```python
employee = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['Jai', 'Princi', 'Gaurav', 'Anuj','Anjali'],
    'department':['HR', 'IT', 'Finance', 'IT','Finance']
})
salary = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[50000,60000,65000,70000,72000],
    'bonus':[5000,6000,6500,7000,7200]
})
```

```python
employee
```

```python
salary
```

```python
pd.merge(employee,salary,on = 'employee_id',how='inner')
```

```python
pd.merge(employee,salary,on = 'employee_id',how='outer')
```

```python
pd.merge(employee,salary,on = 'employee_id',how='left')
```

```python
pd.merge(employee,salary,on = 'employee_id',how='right')
```

**GROUP BY**

```python
data = {
    'Category': ['A','B','A','B','A','B','A','B'],
    'Store': ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales': [100,200,150,250,120,180,200,300],
    'Quantity': [10,15,12,18,8,20,15,25],
    'Date': pd.date_range('2023-01-01', periods = 8)
}
df= pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('Category')
cat
for i,v in cat:
  print(i)
  print(v)
```

```python
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
cat = df.groupby('Store')['Sales'].sum()
cat
```

```python
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

```python
df['Sales'].mean()
```

```python
df['Sales'].agg(['sum','mean','min','max','count','std','median']) #cannot take mode
```

**PIVOT TABLE**

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values = "Sales",index = 'Region',columns="Product",aggfunc = 'median')
```

```python
pivot2= pd.pivot_table(df,values = ["Sales",'Units'],index = 'Region',columns="Product",aggfunc = 'median')
pivot2
```

**CROSS TABS**

```python
pd.crosstab(df['Region'],df['Product'])
```
